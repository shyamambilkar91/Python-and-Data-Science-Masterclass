```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.lineplot(x='sepal_length', y='sepal_width', data = df)
plt.title(" Flower Data ")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.lineplot(x='sepal_length', y='sepal_width', data = df)
plt.xlim(2)
plt.title(" Flower Data ")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

def plot():
    sns.lineplot(x='sepal_length', y='sepal_width', data = df)
    
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    plot()
    
plt.subplot(212)
plot()
```

```python
pallete = sns.color_palette()

sns.palplot(pallete)
plt.show()
```

```python
pallete = sns.color_palette('PiYG',11)

sns.palplot(pallete)
plt.show()
```

```python
pallete = sns.color_palette('Greens',11)

sns.palplot(pallete)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

def plot():
    sns.lineplot(x='sepal_length', y='sepal_width', data = df)

sns.set_palette('vlag')
plt.subplot(211)
plot()
    
sns.set_palette("Accent")
plt.subplot(212)
plot()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.scatterplot(x='sepal_length', y='sepal_width', data = df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.histplot(x='sepal_length', y='sepal_width', data = df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.barplot(x='species', y='sepal_length', data = df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

sns.countplot(x='species', data = df)
plt.show()
```

```python
data = sns.load_dataset('iris')
sns.boxplot(x='species', y='sepal_length', data= data)
plt.show()
```

```python
data = sns.load_dataset('iris')
sns.pairplot(data=data, hue='species')
plt.show()
```

```python
data = sns.load_dataset('iris')
sns.kdeplot(x='sepal_length',y='sepal_width',data= data)
plt.show()
```
