# Pandas -Cleaning Empty Cells

1. Empty cells can be potentially give you wrong result when you analyze data
2. If you want to change the original DataFrame use the inplace = True arguments

## Remove Rows

1. One Way to deal empty cells is to remove that contains empty cells
2. This is usually ok, since data set can be very big, and removing a few rows will not have a big impact on the result

```python
import pandas as pd

df_1 = pd.read_csv('data.csv')
```

```python
df_1.head()
```

```python
df_1.tail()
```

```python
# df_1['Pulse'].unique()
```

```python
new_df = df_1.dropna()
```

```python
new_df
```

```python
df_1
```

**Note: By default dropna() method return a new dataframe, and will not change the original**

```python
df_2 = df_1.copy()
```

```python
df_1.dropna(inplace=True)
```

```python
df_1
```

**Note: Now, the dropna(inplace = True) will NOT return a new DataFrame, but it will remove all rows  containing NULL values from original Datasets**

```python
df_2
```

# Drop Column

```python
df_2 = df_1.drop(['Unnamed: 0'],axis=1)
```

```python
df_2.head()
```

## Replace Empty Cells

1. Another way of dealing with empty cells is to insert a new value insted of deletion of rows
2. This way you do not have to delete entire rows just because of some empty cells
3. The Fillna() method allows us to replace empty cells with a value

```python
df_1['Calories'].fillna(430.4, inplace=True)
```

```python
df_1
```

## Replace Using Imputation Techniques Mean, Median, Mode

1. Common way to replace empty cells is to calculate the mean, mode and median value of the column[feature]
2. Pandas uses the mean(), median(), and mode() methods to calculate the respective values

```python
import pandas as pd
df_3 = pd.read_csv('data.csv')
```

```python
df_3
```

```python
df_4 = df_3['Calories'].mean()
```

```python
df_4
```

```python
df_3['Calories'].fillna(df_4, inplace=True)
```

```python
df_3
```

**Note: Mean - Sum of all values devide by number of values**
