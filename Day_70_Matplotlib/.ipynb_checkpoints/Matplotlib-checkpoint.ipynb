# Matplotlib

1. With the subplot() function you can draw multiple plots in on figure
2. The function subplot() takes three arguments that describe the layout of the figure
3. The Layout is organized in row and column, which represent the first and second arguments
4. The Third argument represent the index of the current plot

#### plt.subplot(1,2,1)
#### The figure has 1 Row, 2 Columns, and this plot is the first plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.array([0,4,5,6,8,7])
y_axis = np.array([0,10,45,63,20,25])

plt.subplot(1,2,1)
plt.plot(x_axis, y_axis)

x_axis = np.array([0,4,6,7,9,10])
y_axis = np.array([0,12,25,35,40,45])

plt.subplot(1, 2, 2)
plt.plot(x_axis, y_axis)

plt.show()
```

```python
x_axis = np.array([0,4,5,6,8,7])
y_axis = np.array([0,10,45,63,20,25])

plt.subplot(2, 1,1)
plt.plot(x_axis, y_axis)

x_axis = np.array([0,4,6,7,9,10])
y_axis = np.array([0,12,25,35,40,45])

plt.subplot(2,1,2)
plt.plot(x_axis, y_axis)

plt.show()
```

```python
x_axis = np.array([0,4,5,6,8,7])
y_axis = np.array([0,10,45,63,20,25])

plt.subplot(1,2,1)
plt.plot(x_axis, y_axis)
plt.title("SALES")

x_axis = np.array([0,4,6,7,9,10])
y_axis = np.array([0,12,25,35,40,45])

plt.subplot(1, 2, 2)
plt.plot(x_axis, y_axis)
plt.title("INCOME")

plt.show()
```

```python
x_axis = np.array([0,4,5,6,8,7])
y_axis = np.array([0,10,45,63,20,25])

plt.subplot(1,2,1)
plt.plot(x_axis, y_axis)
plt.title("SALES")

x_axis = np.array([0,4,6,7,9,10])
y_axis = np.array([0,12,25,35,40,45])

plt.subplot(1, 2, 2)
plt.plot(x_axis, y_axis)
plt.title("INCOME")

plt.suptitle("MY SHOPEE RECORDS")
plt.show()
```

## Scatter Plots

1. with Pyplot, you can use the scatter() function to draw scatter plot
2. The scatter() plot function one dot for each elements (observation). 
3. It needs the array of the same length, one for the value of x-axis and one of the value for y-axis

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.array([5,7,8,7,2,4,17,25,24,20]) 
y_axis = np.array([63,68,72,75,79,88,94,75,63,100])

plt.scatter(x_axis, y_axis)
plt.show()
```

```python
x_axis = np.array([5,7,8,7,2,4,17,25,24,20]) 
y_axis = np.array([63,68,72,75,79,88,94,75,63,100])
plt.scatter(x_axis, y_axis)


x_axis = np.array([7,9,10,9,4,6,19,35,34,30])
y_axis = np.array([[65,78,82,85,89,98,84,65,73,110]])
plt.scatter(x_axis, y_axis)


plt.show()
```

**Note: By Default it will take Blue and Orange color***

```python
x_axis = np.array([5,7,8,7,2,4,17,25,24,20]) 
y_axis = np.array([63,68,72,75,79,88,94,75,63,100])
plt.scatter(x_axis, y_axis, color = 'hotpink')


x_axis = np.array([7,9,10,9,4,6,19,35,34,30])
y_axis = np.array([[65,78,82,85,89,98,84,65,73,110]])
plt.scatter(x_axis, y_axis, color = 'black')


plt.show()
```

# ColorMap

1. The Matplotlib module has a number of available colormap
2. A colormap is like a list of color, where each color value range from o to 100
3. cmap arguments use to assign different colors

```python
x_axis = np.array([5,7,8,7,2,4,17,25,24,20]) 
y_axis = np.array([63,68,72,75,79,88,94,75,63,100])
colors = np.array([0,10,20,40,90,40,50,60,75,85])

plt.scatter(x_axis, y_axis, c = colors, cmap ='viridis')
plt.show()
```

```python
x_axis = np.array([5,7,8,7,2,4,17,25,24,20]) 
y_axis = np.array([63,68,72,75,79,88,94,75,63,100])
colors = np.array([0,10,20,40,90,40,50,60,75,85])
sizes = np.array([0,10,120,40,190,40,50,160,75,85])

plt.scatter(x_axis, y_axis, c = colors, cmap ='viridis', s = sizes)
plt.show()
```

```python
x_axis = np.array([5,7,8,7,2,4,17,25,24,20]) 
y_axis = np.array([63,68,72,75,79,88,94,75,63,100])
colors = np.array([0,10,20,40,90,40,50,60,75,85])
sizes = np.array([0,10,120,40,190,40,50,160,75,85])

plt.scatter(x_axis, y_axis, c = colors, cmap ='viridis', s = sizes, alpha=0.9)
plt.show()
```

# Bar Plot

1. with pyplot you can use the bar() function to draw barplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.array(["Apple","Banana","Cherry","Grapes","Mango","Blackcurrent"])
y_axis = np.array([100,500,1000,200,2000,300])

plt.bar(x_axis, y_axis)
plt.show()
```

```python
x_axis = np.array(["Apple","Banana","Cherry","Grapes","Mango","Blackcurrent"])
y_axis = np.array([100,500,1000,200,2000,300])

plt.bar(x_axis, y_axis, color ='red')
plt.show()
```

```python
x_axis = np.array(["Apple","Banana","Cherry","Grapes","Mango","Blackcurrent"])
y_axis = np.array([100,500,1000,200,2000,300])

plt.bar(x_axis, y_axis, color = 'hotpink')
plt.show()
```

```python
x_axis = np.array(["Apple","Banana","Cherry","Grapes","Mango","Blackcurrent"])
y_axis = np.array([100,500,1000,200,2000,300])

plt.barh(x_axis, y_axis)
plt.show()
```

```python
x_axis = np.array(["Apple","Banana","Cherry","Grapes","Mango","Blackcurrent"])
y_axis = np.array([100,500,1000,200,2000,300])

plt.bar(x_axis, y_axis, width=0.9)
plt.show()
```

```python
x_axis = np.array(["Apple","Banana","Cherry","Grapes","Mango","Blackcurrent"])
y_axis = np.array([100,500,1000,200,2000,300])

plt.barh(x_axis, y_axis, height= 0.5)
plt.show()
```

# Pie Chart

1. with pyplot you can use the pie() function to draw pie chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([45,62,57,36,20])

plt.pie(x)
plt.show()
```

```python
x = np.array([45,62,57,36,20])
my_lables = ["Apple",'Mango','Cherry','Blackcurrent','Grapes']

plt.pie(x, labels= my_lables)
plt.show()
```

```python
x = np.array([45,62,57,36,20])
my_lables = ["Apple",'Mango','Cherry','Blackcurrent','Grapes']

plt.pie(x, labels= my_lables, startangle= 90)
plt.show()
```

```python
x = np.array([45,62,57,36,20])
my_lables = ["Apple",'Mango','Cherry','Blackcurrent','Grapes']
my_explode = [0,0,0.2,0,0]

plt.pie(x, labels= my_lables, startangle= 90, explode =my_explode)
plt.show()
```

```python
x = np.array([45,62,57,36,20])
my_lables = ["Apple",'Mango','Cherry','Blackcurrent','Grapes']
my_explode = [0,0,0.2,0,0]

plt.pie(x, labels= my_lables, startangle= 90, explode =my_explode, shadow= True)
plt.show()
```

```python
x = np.array([45,62,57,36,20])
my_lables = ["Apple",'Mango','Cherry','Blackcurrent','Grapes']
my_explode = [0,0,0.2,0,0]
my_color = ['red','orange','green','black','hotpink']

plt.pie(x, labels= my_lables, startangle= 90, explode =my_explode, shadow= True, colors = my_color)
plt.show()
```

```python
x = np.array([45,62,57,36,20])
my_lables = ["Apple",'Mango','Cherry','Blackcurrent','Grapes']
my_explode = [0,0,0.2,0,0]
my_color = ['red','orange','green','black','hotpink']

plt.pie(x, labels= my_lables, startangle= 90, explode =my_explode, shadow= True, colors = my_color)
plt.legend(title = "Fruits")
plt.show()
```
