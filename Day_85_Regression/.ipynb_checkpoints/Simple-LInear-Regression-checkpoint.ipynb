## Feature Engineering

1. Trying different transformation of data to estimate normal distribution and to remove skewness

```python
sns.distplot(df_1['delivery_time'], bins = 10, kde = True)
plt.title("Before Transformation")
sns.displot(np.log(df_1['delivery_time']), bins= 10, kde= True)
plt.title("After Transformation")
plt.show()
```

```python
labels = ['Before Transformation', 'After Transformation']
sns.distplot(df_1['delivery_time'], bins= 10, kde=True)
sns.distplot(np.log(df_1['delivery_time']), bins=10, kde=True)
plt.legend(labels)
plt.show()
```

1. Log Transformation affects the data as it scaled the values down
2. Before Prediction it is necessory to reverse scaled the data

```python
smf.qqplot(df_1['delivery_time'], line= 'r')
plt.title("No Transformation")

smf.qqplot(np.log(df_1['delivery_time']), line= 'r')
plt.title('Log Transformation')

smf.qqplot(np.sqrt(df_1['delivery_time']), line= 'r')
plt.title('Square Root Transformation')

smf.qqplot(np.cbrt(df_1['delivery_time']), line= 'r')
plt.title('Cube Root Transformation')
plt.show()
```

```python
labels = ['Before Transformation', 'After Transformation']
sns.distplot(df_1['sorting_time'], bins= 10, kde=True)
sns.distplot(np.log(df_1['sorting_time']), bins=10, kde=True)
plt.legend(labels)
plt.show()
```

```python
smf.qqplot(df_1['sorting_time'], line= 'r')
plt.title("No Transformation")

smf.qqplot(np.log(df_1['sorting_time']), line= 'r')
plt.title('Log Transformation')

smf.qqplot(np.sqrt(df_1['sorting_time']), line= 'r')
plt.title('Square Root Transformation')

smf.qqplot(np.cbrt(df_1['sorting_time']), line= 'r')
plt.title('Cube Root Transformation')
plt.show()
```

**Note: We only perform any data transformation when data is skewed or NOt Normal**

### Fitting Linear Regression Model

1. Using Ordinary least square regression (ols)
2. It is statistical method of analysis that estimate the relationship between  one or more indedependent variable and a depedent variable
3. The method estimate relationship by minimizing the sum of square of the difference between th observed data and predicted value of the dependent variable configured as straight line

```python
model = sm.ols('delivery_time ~ sorting_time', data=df_1).fit()
```

```python
model.summary()
```

1. The R-Squared and Adjusted Score are still below o.8, try some another transformation to get score higher

### Log Transformation

```python
model_2 = sm.ols('np.log(delivery_time) ~ np.log(sorting_time)', data=df_1).fit()
```

```python
model_2.summary()
```

## Model Testing

Y = Beta0 + Beta1 * (x)

```python
model_2.params
```

1. Intercept (Beta0) value = 1.74
2. np.log(Beta1) value = 0.5
3. Hypothesis Testing of x variable by finding test_stat and P_value for Beta1 (P_value <0.5 Reject Null)
4. Null Hypothesis as Beta1 = 0 (No slope) and Alternate Hypethesis as Beta1 = 0 (some or significant slop)

```python
print(model.tvalues,'\n',model.pvalues)
```

```python
model_2.rsquared, model_2.rsquared_adj
```

```python
import statsmodels.api as sm
sm.qqplot(model.resid, line='q')
plt.title('Normal Q-Q plot of Residual of model without any data transformation')
plt.show()
```

```python
import statsmodels.api as sm
sm.qqplot(model_2.resid, line='q')
plt.title('Normal Q-Q plot of Residual of model with data transformation')
plt.show()
```

### Residual plot to check Homoscedasticity or Hetroscedasticity

```python
def get_standardized_values(vals):
    return (vals - vals.mean()) / vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues), get_standardized_values(model.resid))
plt.title('Normal Q-Q plot of Residual of model without any data transformation')
plt.xlabel('Standardized fitted values')
plt.ylabel('Standard residual values')
plt.show()
```

```python
plt.scatter(get_standardized_values(model_2.fittedvalues), get_standardized_values(model_2.resid))
plt.title('Normal Q-Q plot of Residual of model with log data transformation')
plt.xlabel('Standardized fitted values')
plt.ylabel('Standard residual values')
plt.show()
```
