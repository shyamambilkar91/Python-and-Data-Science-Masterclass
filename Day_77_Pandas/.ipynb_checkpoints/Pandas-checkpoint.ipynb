7. Free to Download and use

```python
my_db = sql.connect(host = '127.0.0.1', 
                    port = 3308, 
                    user = 'root', 
                    passwd = 'root', 
                    database = 'shopee',
                    auth_plugin = 'mysql_native_password'
                   )
```
