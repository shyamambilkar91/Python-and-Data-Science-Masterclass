# Data Distribution

## 1. Normal Data Distribution

1. The Normal Data Distribution is one of the most important Data Distribution
2. It is also called as Gaussian Distribution
3. It fits probabibility distribution of many events Example - IQ Scores, HeartBeats etc.
4. Use the random.normal() methods to get Normal Data Distribution
5. It has required three Parameters:
    1. loc - [ Mean ] Where the peak of the bell exist
    2. scale - [ Standard Deviation ] how flat the graph distribution should be
    3. size - The shape of the returned array

### 1. Generate a Random Normal Distribution of size 2 * 3

```python
from numpy import random

normal_dist = random.normal(size=(2,3))
print(normal_dist)
```

### 2. Generate a Random Normal Distribution of size 2 * 3 with mean of 1 and standard deviation 2

```python
from numpy import random

normal_dist = random.normal(loc=1, scale= 2, size=(2, 3))
print(normal_dist)
```

### 3. Visualization of Normal Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist = False)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist = True)
plt.show()
```

**Note:The curve a Normal Distribution it is also called a Bell Curve because of the bell shape curve**

# 2. Binomial Distribution

1. Binomial Distribution also called Discrete Distribution
2. It describe the outcome of Binary scenario Example - Toss of coin(either it should be Head or either it should be Tell)
3. It has three Parameters:
    1. n - number of trials
    2. p - probability of occurane of binary scenario
    3. size - The shape of returned array

**Discrete Distribution: The distribution is defined at separate set of events**

```python
from numpy import random

binomial_dist = random.binomial(n = 10, p=0.5,size=10)
print(binomial_dist)
```

### Visualization of Binomial distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n =10, p=0.5, size=1000), hist=True, kde=False)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n =10, p=0.5, size=1000), hist=True, kde=True)
plt.show()
```

# 3. Difference Between Normal and Binomial Distribution

1. Biggest Difference is that is Normal distribution continous whereas Binomial is discrete
2. But if there are enough data points it will be quite similar to normal distribtution with certain loc and scale

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc= 50, scale= 5, size= 1000),hist= False, label="Normal distribution")
sns.distplot(random.binomial(n = 100, p=0.5, size=1000),hist=False, label="Binomial distribution")

plt.show()
```

## 3. Poission Distribution

1. Poisson Distribution is a Descrete Distribution
2. It estimate how many times an event can happen in a specified time
3. Example:
    1. If someone eats twice a day what is the probability he will eat thrice ?
    2. It has two parameters:
        1. lam - rate or known number of occurances 
        2. size - The shape of the returned array

```python
from numpy import random

poision_dist = random.poisson(lam=2, size=10)
print(poision_dist)
```

### Visualization of Poisson Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.poisson(lam=2, size=1000), kde=False)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.poisson(lam=2, size=1000), kde=True)
plt.show()
```

## Difference Between Normal Distribution and Poission Distribution

1. Normal Distribution is contineous Poission is descrete
2. But we can see that similar to binomial for a large enough poission distribution it will becomes similar to normal distribution with certain std dev and mean

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50,scale=7, size=1000),hist=False, label='normal')
sns.distplot(random.poisson(lam=50, size=1000),hist=False, label='poission')

plt.show()
```
