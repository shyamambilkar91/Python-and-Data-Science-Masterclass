```python
import matplotlib.pyplot as plt
import numpy as np


x_axis = np.array([0,6])
y_axis = np.array([0,60])

plt.plot(x_axis, y_axis, marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x_axis = np.array([0,6])
y_axis = np.array([0,60])

plt.plot(x_axis, y_axis, "o")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x_axis = np.array([0,6,12,18,20])
y_axis = np.array([0,60,45,63,85])

plt.plot(x_axis, y_axis, "o")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x_axis = np.array([0,6,12,18,20])
y_axis = np.array([0,60,45,63,95])

plt.plot(x_axis, y_axis, marker = "o")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x_axis = np.array([0,6,12,18,20])
y_axis = np.array([0,60,45,63,95])

plt.plot(x_axis, y_axis, marker = "*")
plt.show()
```
