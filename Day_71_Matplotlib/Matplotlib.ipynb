# Matplotlib

## 1. Steam Plot In Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.linspace(0.1, 2 * np.pi, 41)
y_axis = np.exp(np.sin(x_axis))

plt.stem(x_axis, y_axis, use_line_collection= True)
plt.show()
```

```python
x_axis = np.linspace(0.1, 2 * np.pi, 41)
y_axis = np.exp(np.sin(x_axis))

plt.stem(x_axis, y_axis, use_line_collection= False)
plt.show()
```

## 2. Stack Plot in Matplotlib

```python
import matplotlib.pyplot as plt

num_of_days = [1, 2, 3, 4, 5]
no_of_days_studying = [7, 8, 6, 11, 7]
no_of_days_playing = [8, 5, 7, 8, 13]

plt.stackplot(num_of_days, no_of_days_studying, no_of_days_playing, colors=['r','g','b'])

plt.xlabel("No of Days")
plt.ylabel("No of Hours")
plt.title("Representation of Study and Playing wrt to Days")

plt.show()
```

## 3. Box Plot in Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize=(10,7))

plt.boxplot(data)
plt.show()
```

## 4. Error Plot in Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x_axis = [1, 2, 3, 4, 5, 6, 7]
y_axis = [1, 2, 1, 2, 1, 2, 1]

y_error = 0.3

plt.plot(x_axis, y_axis)
plt.errorbar(x_axis, y_axis, yerr=y_error, fmt='o')
```

## 5. Violin Plot in Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
uniform_array = np.arange(-100, 100)

normal = np.random.normal(size = 100) * 30
# print(normal)

fig, (ax1, ax2) = plt.subplots(nrows = 1, 
            ncols = 2,
            figsize = (9, 4), 
            sharey = True)

ax1.set_title("Uniform Distribution")
ax1.set_ylabel("Observed Value")
ax1.violinplot(uniform_array)

ax2.set_title("Normal Distribution")
ax2.violinplot(normal)

plt.show()
```
