### Calculate the Entropy of Temprature[Hot, Cool, Mild]

```python
1. P[Temp -YES] = 2
2. N[Temp - NO] = 2
3. I(P,N) = (2,2)

I(P,N) = I(2,2)
       = -(2 / (2 + 2)LOG2(2 /(2 + 2)) - (2 /(2 + 2)LOG2( 2 /(2 + 2))))
       = -((2/4)log2(2/4)) - ((2 / 4)log2(2 / 4))
       = -(-0.5) * (-1) + (-0.5) * (-1)
       = (0.5) + (0.5)
       = 1
```

### Compute the Entropy for Humadity (High, Normal)

```python
IG(P_i, N_i) =  (0, 2)
             = -(0 /(0 + 2) log2 (0 / 0 + 2)) - (2 / (2 + 0) log2 (2 /2 + 0))
             = -(0/ 2) log2(0 / 2) - (2 /2) log2(2/2)
             =  0 - 0
             =  0
```

```python
Hence Gain(Humadity) = I(P,N) - E(Temprature)
                       = 0.971 - 0 = 0.971
```

### Compute the Entropy for Humadity (Weak, Strong)

```python
P[STRONG] = [1,1] = 2 / 5
P[WEAK] = [1, 2] = 3 / 5


E(WIND)       = (3/5) * (1,2) + (2/5) * I(1, 1)
              = 0.66 * 0.91 + 0.4 * 1
              = 0.60 + 0.4
              = 0.9508
```

```python
Hence Gain(WIND) = I(P,N) - E(Temprature)
                       = 0.971 - 0.950 = 0.02
```
