# Matplotlib

## Markers

1. [ 'o' ] - Circle
2. [ '*' ] - Star
3. [ "." ] - Point
4. [ "," ] - Pixel
5. [ "x" ] - X
6. [ 'X' ] - X(filled)
7. [ '+' ] - Plus
8. [ 'P' ] - Plus (filled)
9. [ 's' ] - Square
10. [ 'D' ] - Diamond
11. [ 'd' ] - Diamond (thin)
12. [ 'p' ] - Pentagon
13. [ 'h' ] - Hexagon
14. [ 'H' ] - Hexagon
15. [ '<' ] - Triangle Left
16. [ '>' ] - Triangle Right
17. [ '^' ] - Triangle Up
18. [ 'v' ] - Triangle Down
19. [ '|' ] - Vline
20. [ '_' ] - Hline

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])


plt.plot(x_axis, y_axis)
plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])


plt.plot(x_axis, y_axis, marker = 'o')
plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])


plt.plot(x_axis, y_axis, 'd')
plt.show()
```

## Format String [fmt]

1. You can use shortcut string notation parameter to specify the marker
2. This parameter also called as **fmt**
3. syntax:
    1. marker|line|color [:]

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, 'o:r')
plt.show()
```

## Line Reference

1. Solid Line ['-']
2. Dotted Line ['.']
3. Dashed Line ['--']
4. Dashed/Dotted Line ['-.']

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, 'o-', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, 'o--', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, 'o-.', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, 'o:', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

**Note: If you leave out the line value in fmt parameter, no line will be plotted**

## Color Reference

1. [ 'r' ] = Red
2. [ 'g' ] = Green
3. [ 'b' ] = Blue
4. [ 'c' ] = Cyan
5. [ 'm' ] = Magneta
6. [ 'y' ] = Yellow
7. [ 'k' ] = Black
8. [ 'w' ] = White

## Marker Size

1. You can use parameter markersize or shorter version of this is ms

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, marker ='o', ms = 20)
plt.show()
```

## Marker Color

1. You can use parameter **markeredgecolor** or shorter version of this is **mec**

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, marker ='o', ms = 20, mec = 'r')
plt.show()
```

## Marker Face Color

1. You can use parameter **markerfacecolor** or shorter version of this is **mfc**

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, marker ='o', ms = 20, mec = 'r', mfc = 'green')
plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, marker ='o', ms = 20, mec = 'hotpink', mfc = 'hotpink')
plt.show()
```

## Line Style

1. You can use parameter **linestyle** or shorter version of this is ls

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

## Line Width

1. You can use parameter **linewidth** or shorter version of this is lw

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed',lw= '5', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

## Multiple Lines 

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

x_axis_1 = np.array([0,5,10,15,20,25,30])
y_axis_1 = np.array([0,7,13,18,24,56,78])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', mec = 'red', mfc = 'red')
plt.plot(x_axis_1, y_axis_1, marker ='o',linestyle = 'dashed', mec = 'green', mfc = 'green')
plt.show()
```

## Matplotlib labels and Title

1. with Pyplot you can use:
    1. xlabel()
    2. ylabel()

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

x_axis_1 = np.array([0,5,10,15,20,25,30])
y_axis_1 = np.array([0,7,13,18,24,56,78])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', mec = 'red', mfc = 'red')
plt.plot(x_axis_1, y_axis_1, marker ='o',linestyle = 'dashed', mec = 'green', mfc = 'green')

plt.xlabel("Average Value")
plt.ylabel("Calories Burned")

plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

x_axis_1 = np.array([0,5,10,15,20,25,30])
y_axis_1 = np.array([0,7,13,18,24,56,78])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', mec = 'red', mfc = 'red')
plt.plot(x_axis_1, y_axis_1, marker ='o',linestyle = 'dashed', mec = 'green', mfc = 'green')

plt.title("Sports Related Data")
plt.xlabel("Average Value")
plt.ylabel("Calories Burned")

plt.show()
```

## Set Font Properties for Title and Labels

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

x_axis_1 = np.array([0,5,10,15,20,25,30])
y_axis_1 = np.array([0,7,13,18,24,56,78])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', mec = 'red', mfc = 'red')
plt.plot(x_axis_1, y_axis_1, marker ='o',linestyle = 'dashed', mec = 'green', mfc = 'green')

font_1 = {'family':'serif','color':'blue','size':20}
font_2 = {'family':'serif','color':'green','size':15}

plt.title("Sports Related Data", fontdict=font_1)
plt.xlabel("Average Value", fontdict= font_2)
plt.ylabel("Calories Burned", fontdict=font_2)

plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

x_axis_1 = np.array([0,5,10,15,20,25,30])
y_axis_1 = np.array([0,7,13,18,24,56,78])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', mec = 'red', mfc = 'red')
plt.plot(x_axis_1, y_axis_1, marker ='o',linestyle = 'dashed', mec = 'green', mfc = 'green')

font_1 = {'family':'serif','color':'blue','size':20}
font_2 = {'family':'serif','color':'green','size':15}

plt.title("Sports Related Data", fontdict=font_1, loc='right')
plt.xlabel("Average Value", fontdict= font_2)
plt.ylabel("Calories Burned", fontdict=font_2)

plt.show()
```

## Add Grid Lines to a Plot

1. You can use function name called called as grid()

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

x_axis_1 = np.array([0,5,10,15,20,25,30])
y_axis_1 = np.array([0,7,13,18,24,56,78])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', mec = 'red', mfc = 'red')
plt.plot(x_axis_1, y_axis_1, marker ='o',linestyle = 'dashed', mec = 'green', mfc = 'green')

font_1 = {'family':'serif','color':'blue','size':20}
font_2 = {'family':'serif','color':'green','size':15}

plt.title("Sports Related Data", fontdict=font_1, loc='right')
plt.xlabel("Average Value", fontdict= font_2)
plt.ylabel("Calories Burned", fontdict=font_2)

plt.grid()
plt.show()
```

```python
x_axis = np.array([10,20,30,40,50,45,63])
y_axis = np.array([0,5,43,10,12,75,45])

x_axis_1 = np.array([0,5,10,15,20,25,30])
y_axis_1 = np.array([0,7,13,18,24,56,78])

plt.plot(x_axis, y_axis, marker ='o',linestyle = 'dashed', mec = 'red', mfc = 'red')
plt.plot(x_axis_1, y_axis_1, marker ='o',linestyle = 'dashed', mec = 'green', mfc = 'green')

font_1 = {'family':'serif','color':'blue','size':20}
font_2 = {'family':'serif','color':'green','size':15}

plt.title("Sports Related Data", fontdict=font_1, loc='right')
plt.xlabel("Average Value", fontdict= font_2)
plt.ylabel("Calories Burned", fontdict=font_2)

plt.grid(axis='x')
plt.show()
```
