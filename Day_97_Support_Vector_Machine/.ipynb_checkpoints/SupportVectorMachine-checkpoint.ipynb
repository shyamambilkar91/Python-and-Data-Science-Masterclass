# Support Vector Machine

1. **Linear Kernal** - used when the data is linearly separable
2. **Polynomial Kernal** - Used when the data is not linearly separable but can be separated with a polynomial boundary
3. **Radial Basis Function (RBF) Kernal** - A Popular Kernal that maps data to an infinite dimensional. It's useful for non-linear
4. **Sigmoid Kernal** - Often used in Neural Network

### Independent and Dependent Variable

```python
x = df_2.drop('cancer', axis=1)
```

```python
y = df_2['cancer']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=True)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model
```

```python
model.fit(x_train, y_train)
```

## Prediction

```python
pred = model.predict(x_test)
```

```python
pred
```

```python
pred[:10]
```

```python
y_test[:10]
```

### Evalution 

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve,accuracy_score,mean_squared_error
```

```python
conf_matrix = confusion_matrix(y_test, pred)
```

```python
sns.heatmap(conf_matrix, annot=True)
```

```python
conf_matrix
```

```python
print(classification_report(y_test, pred))
```

```python
print("Accuracy of SVC are: ", accuracy_score(y_test, pred))
```

```python
mean_squared_error(y_test, pred)
```
