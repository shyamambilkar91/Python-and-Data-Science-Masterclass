## Independent and Dependent Variable

```python
x = df.drop('Outcome',axis=1)
```

```python
y = df['Outcome']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=True)
```

```python
x_train[:5]
```

```python
y_train.head()
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
col_names = x_train.columns
print(col_names)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=10)
```

```python
decision_model = model.fit(x_train, y_train)
```

```python
model = DecisionTreeClassifier(criterion='gini', max_depth= 6)
```

```python
decision_model = model.fit(x_train, y_train)
```

```python
prediction = decision_model.predict(x_test)
```

```python
print(prediction)
```

```python
print(y_test)
```

```python
prediction_train = decision_model.predict(x_train)
```

```python
print(prediction_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
train_acc = accuracy_score(y_train, prediction_train)
```

```python
print(train_acc)
```

```python
test_acc = accuracy_score(y_test, prediction)
```

```python
print(test_acc)
```

```python
conf_matrix = confusion_matrix(y_test, prediction)
```

```python
conf_matrix
```

```python
sns.heatmap(conf_matrix, annot=True)
```

```python
print(classification_report(y_test, prediction))
```
